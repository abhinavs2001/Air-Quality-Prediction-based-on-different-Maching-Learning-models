# Importing all the necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('file1.csv')   # reading the csv file
df   
```

# Creating Model for predicting the Output

```python
final_df= df[['AQI', 'AQI_Bucket']].copy()
final_df
```

# Labelling the categories.

```python
final_df['AQI_Bucket'] = final_df['AQI_Bucket'].map({'Good' :0, 'Satisfactory' :1, 'Moderate' :2, 'Poor' :3, 'Very Poor' :4, 'Severe' :5}).astype(int) #mapping numbers
final_df.head()
```

# Predicting the values of AQI_Bucket w.r.t values of AQI using Linear Regression

```python
X = final_df[['AQI']]
y = final_df[['AQI_Bucket']]
```

```python
# division into training and test sets.
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0) 
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()  
regressor.fit(X_train, y_train) 
y_pred = regressor.predict(X_test)

print("Training complete.")
```

# Taking AQI as input for finding the category

```python
import math  
print("Enter the value of AQI:")
AQI = float(input("AQI : "))
output = regressor.predict([[AQI]])
print( math.ceil(output))
#0-->Good
#1-->Satisfactory
#2-->moderate
#3-->poor
#4-->Very poor
#5-->Severe
```

```python
y_pre=np.ceil(y_pred)
y_pre=pd.DataFrame(y_pre)
```
